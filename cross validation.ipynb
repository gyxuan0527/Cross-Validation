```python
data['workclass'] = [data['workclass'].value_counts().idxmax() if x==' ?' else x for x in data['workclass']]
data= pd.get_dummies(data=data,columns=['workclass'])

data= pd.get_dummies(data=data,columns=['education'])

data= pd.get_dummies(data=data,columns=['marital_status'])
data['occupation'] = [data['occupation'].value_counts().idxmax() if x==' ?' else x for x in data['occupation']]
data= pd.get_dummies(data=data,columns=['occupation'])

data= pd.get_dummies(data=data,columns=['relationship'])

data= pd.get_dummies(data=data,columns=['race'])

data= pd.get_dummies(data=data,columns=['sex'])
data['native_country'] = [data['native_country'].value_counts().idxmax() if x==' ?' else x for x in data['native_country']]
data= pd.get_dummies(data=data,columns=['native_country'])
pd.DataFrame(data)
```

```python
y=data['income']
data=data.drop(labels=['income'],axis='columns')
y
```

```python

def K_fold_CV(k, data):
    subset_size=round(len(data)/k)
    Accuracy=0
    for i in range(k):
        x_test=data[i*subset_size:min((i+1)*subset_size,len(data))]
        y_test=y[i*subset_size:min((i+1)*subset_size,len(data))]
        if (i+1)*subset_size==len(data):
            x_train=data[0:i*subset_size]
            y_train=y[0:i*subset_size]
        elif i==0:
            x_train=data[(i+1)*subset_size:]
            y_train=y[(i+1)*subset_size:]
        else:
            x_train=data[(i+1)*subset_size:].append(data[0:i*subset_size])
            y_train=y[(i+1)*subset_size:].append(y[0:i*subset_size])

        forest = RandomForestClassifier(n_estimators = 100)
        forest_fit = forest.fit(x_train, y_train)
        
        y_pred = forest.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        Accuracy=Accuracy+accuracy
    return Accuracy/k
```

```python
K_fold_CV(10,data)
```
